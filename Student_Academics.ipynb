```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in the csv file as a pandas dataframe
student = pd.read_csv("Study_Hour.csv")
```

```python
#explore data
student.info()
```

```python
student.shape
```

```python
student
```

```python
# count the unique values in the Hours column
student.Hours.value_counts()
```

```python
# count the unique values in the Score column
student.Scores.value_counts()
```

```python
#customizing the column names as per need
new_cols = {
    "Hours":"Study_Hours",
    "Scores":"Student_Score"
}

#renaming the columns
student.rename(new_cols,axis=1,inplace=True)
```

```python
student
```

```python
# changing the Student_Score to percentage (i.e. a value between 0 and 1)
student.Student_Score = student.Student_Score/100
```

```python
student.head()
```

```python
#describing dataset
student.describe()
```

```python
plt.hist(student["Study_Hours"],bins=10,edgecolor="y",color="m")
plt.title("Student Growth")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show
```

```python
# Calculate correlation coefficients between numerical features
correlation_matrix = student.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Visualize correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```
